## Doc string 

```python
def name_function(name):
    
    """
    DOCSTRING : (info about the function ) This function prints the name of the person
    
    INPUT : Name as string
    
    OUTPUT : "Hello" + Name as string 
    
    """
    print('Hello...' + name )
```

## One more simple function

```python
# create a function that returns squared value of the input

def square_num(number):
    """
    
    DOCSTRING : This function returns the squared value of the input
    
    INPUT : Single Integer or float value 
    
    OUTPUT : Squared Integer or float value 
    
    """
    
    return (number*number)
```

```python
help(square_num)
```

```python
square_num(9)
```

## Text Function 

```python
#create a function that finds out if input word is in given string 

def word_check(sentence,word):
    """
    
    DOCSTRING : This function returns True or False in case input word present in the sentence
    
    INPUT : Sentence and word that needs to be validated to be present 
    
    OUTPUT : Boolean Values ( TRUE or FALSE)
    
    """
    
    if word in sentence:
        return True 
    else :
        return False
```

```python
word_check('This is a text function to validate presence of word python','python')
```

# Arguments (**args) & Keyword Arguments ( **kwargs)

```python
# **args
# **kwargs
```

```python
#create simple function with positional arguments 

def five_percent(a,b):
    
    # a,b are called positional arguments 
    """
    
    DOCSTRING : This function returns 5 % value of total sum of input
    
    INPUT : Integer values
    
    OUTPUT :  5 % of total input values
    
    """
    val=sum((a,b))*0.05
    
    return val
    
    
```

```python
five_percent(1000,750)
#1000 is assigned to a because it was at the first positon and 750 assigned to b becuase it was at second position in function 
```

#### Incase where we are not sure of the number of arguments/parameters in advanvce , we use args & kwargs
#### It allows us to take arbitary number of arguments with out declaring in advance


```python
#create simple function with arbitary arguments 

def five_percent(*args):
    
    
    """
    
    DOCSTRING : This function returns 5 % value of total sum of input
    
    INPUT : Integer values
    
    OUTPUT :  5 % of total input values
    
    """
    val=sum((args))*0.05
    
    return val
    
```

```python
#use the function by passing any number of arguments
five_percent(300,500,100,400)
```

```python
#### one case use any keyword insteasd of args as well , but it is used widely by most of the developer community

#create simple function with arbitary arguments 

def five_percent(*london):
    
    
    """
    
    DOCSTRING : This function returns 5 % value of total sum of input
    
    INPUT : Integer values
    
    OUTPUT :  5 % of total input values
    
    """
    val=sum((london))*0.05
    
    return val
```

```python
five_percent(33,55,66,77,88)
```

## Keyword Arguments

```python
#In the case when the input parameters are key value pairs /dictionary

def capital(*args,**kwargs):
    
    """
    
    DOCSTRING : This function returns capital of the country
    
    INPUT : Key value pairs of Country and Capital 
    
    OUTPUT :  Capital of the country
    
    """
    
    
    
    print("The capital of {} is {}".format(args[0],kwargs['India']))
    
    
```

```python
capital('India','UK','France',India='Delhi',UK='London',France='Paris')
```
