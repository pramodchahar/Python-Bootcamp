```python
#create a sample string 
test_string=['Roger','Nadal','Djokovic','Sharapova','Williams','Murray']


#create an empty list 
new_string=[]

#Create a simple for loop to iterate over the list items and append new list

for item in test_string:
    new_string.append(item)

#validate the new list items
print(new_string)
```

```python
# Replicating the same functionalty using List Comprehension
lc_new_string=[item for item in test_string]

print(lc_new_string)
```
